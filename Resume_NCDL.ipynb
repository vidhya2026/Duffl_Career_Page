```python
#Import the neccessary
import os, re, tempfile
import pdfplumber
import spacy
import pandas as pd
from docx import Document
from datetime import datetime
from sentence_transformers import SentenceTransformer, util
from transformers import pipeline
from flask import Flask, request, jsonify
from pyngrok import ngrok
from google.colab import files
from IPython.display import display, clear_output
import warnings
warnings.filterwarnings("ignore")

print("Loading models...")
embedder   = SentenceTransformer('all-MiniLM-L6-v2')
ner_pipe   = pipeline("ner", model="dslim/bert-base-NER",    aggregation_strategy="simple")
ner_large  = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", aggregation_strategy="simple")
nlp        = spacy.load("en_core_web_sm")
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
print("Models loaded!")
```

```python
#  TEXT EXTRACTION

def extract_text_pdf(path):
    full, left, right = "", "", ""
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            w = page.width
            h = page.height

            # Get actual page bounds (handles non-zero origin PDFs)
            x0 = page.bbox[0]
            y0 = page.bbox[1]
            x1 = page.bbox[2]
            y1 = page.bbox[3]

            t = page.extract_text()
            if t: full += t + "\n"

            # Clamp crop coords within actual page bbox
            mid_x = x0 + (x1 - x0) * 0.42

            try:
                lt = page.crop((x0, y0, mid_x, y1)).extract_text()
                if lt: left += lt + "\n"
            except Exception:
                pass

            try:
                rt = page.crop((mid_x, y0, x1, y1)).extract_text()
                if rt: right += rt + "\n"
            except Exception:
                pass

    return full.strip(), left.strip(), right.strip()

def extract_text_docx(path):
    doc  = Document(path)
    text = "\n".join([p.text for p in doc.paragraphs if p.text.strip()])
    for table in doc.tables:
        for row in table.rows:
            for cell in row.cells:
                text += "\n" + cell.text
    return text.strip(), "", ""

def extract_text(path):
    if path.lower().endswith(".pdf"):  return extract_text_pdf(path)
    if path.lower().endswith(".docx"): return extract_text_docx(path)
    if path.lower().endswith(".txt"):
        with open(path, "r", errors="ignore") as f: c = f.read()
        return c, "", ""
    return "", "", ""
```

```python
#   NAME EXTRACTION

INVALID_WORDS = {
    # portals
    "naukri","indeed","linkedin","monster","shine","foundit",
    "glassdoor","hirist","instahyre","apna","freshersworld",
    "timesjobs","workindia","internshala","iimjobs",
    # roles / tech
    "ui","ux","uiux","uxui","dev","devops","qa","ba","pm",
    "frontend","backend","fullstack","full","stack","data",
    "cloud","ml","ai","hr","it","ios","web","sde","swe",
    "react","angular","vue","java","python","dotnet","node",
    "flutter","android","php","ruby","golang","rust",
    "aws","azure","gcp","docker","kubernetes",
    # resume section words
    "resume","curriculum","vitae","profile","contact","summary",
    "objective","experience","education","skills","address","phone",
    "email","mobile","github","portfolio","work","professional",
    "personal","details","information","career","about","project",
    "references","developer","engineer","manager","analyst",
    "designer","consultant","specialist","lead","head","senior",
    "junior","intern","fresher","candidate","applicant",
    # file/doc words
    "updated","new","final","cv","doc","copy","sample","test",
    "example","demo","blank","template","format","latest","hire",
    "upload","document","file","draft","version",
}


# ── FILENAME  (tried FIRST) ───────────────────────────────────

def get_name_from_filename(filepath):
    """
    For Naukri-style:  Naukri_SIVARAMANV[4y_0m].pdf
    Steps:
      1. Strip extension
      2. Remove bracket/paren blocks  e.g. [4y_0m]  (2y_2m)
      3. Split on _ - . space
      4. Discard tokens that are portal names, all-digits, or
         contain digits (like 4y, 0m, 2y)
      5. Whatever tokens remain → that is the candidate name
      6. Must have at least 1 token and each token >= 2 chars
    """
    raw = os.path.splitext(os.path.basename(filepath))[0]

    # remove bracket / paren blocks
    raw = re.sub(r'\[.*?\]', ' ', raw)
    raw = re.sub(r'\(.*?\)', ' ', raw)

    # split camelCase
    raw = re.sub(r'([a-z])([A-Z])', r'\1 \2', raw)

    # normalise separators
    raw = re.sub(r'[_\-\.\s]+', ' ', raw).strip()

    tokens = raw.split()
    good = []
    for tok in tokens:
        tl = tok.lower()
        if not tok:                             continue
        if tl in INVALID_WORDS:                continue  # skip portal/role words
        if any(c.isdigit() for c in tok):      continue  # skip anything with digits
        if re.search(r'[^A-Za-z]', tok):       continue  # letters only
        if len(tok) < 2:                        continue
        good.append(tok.capitalize())

    if not good:
        return None

    return " ".join(good)          # return ALL remaining tokens as the name


# ── NER helpers ───────────────────────────────────────────────

def get_name_bert(text, model):
    try:
        entities = model(text[:700])
        names = [e['word'] for e in entities if e['entity_group'] == 'PER']
        if names:
            raw = " ".join(names[:4]).replace(" ##", "").strip()
            parts = [w for w in raw.split()
                     if w.lower() not in INVALID_WORDS
                     and not any(c.isdigit() for c in w)
                     and len(w) >= 2]
            if len(parts) >= 2:
                return " ".join(parts[:4])
    except:
        pass
    return None

def get_name_spacy(text):
    try:
        doc = nlp(text[:700])
        for ent in doc.ents:
            if ent.label_ == "PERSON":
                candidate = ent.text.strip()
                words = candidate.split()
                if (2 <= len(words) <= 5
                        and all(not any(c.isdigit() for c in w) for w in words)
                        and all(w.lower() not in INVALID_WORDS for w in words)):
                    return candidate
    except:
        pass
    return None

_TITLE_NAME_RE = re.compile(
    r'\b([A-Z][a-z]{1,20}(?:\s+[A-Z][a-z]{1,20}){1,3})\b'
)

def get_name_regex(text):
    top = "\n".join(text.splitlines()[:12])
    for m in _TITLE_NAME_RE.findall(top):
        words = m.split()
        if all(w.lower() not in INVALID_WORDS for w in words):
            return m
    return None


# ── MASTER EXTRACTOR  (filename is tried FIRST) ───────────────

def extract_name(full_text, left_col, right_col, filepath):
    """
    1. Filename  — always tried first
    2. BERT-base  on top 10 lines
    3. BERT-large on top 10 lines
    4. spaCy      on top 10 lines
    5. BERT / spaCy on full text
    6. Regex Title-Case on top lines
    7. "Name Not Found"
    """

    # ── 1. FILENAME ──────────────────────────────────────────
    fname_result = get_name_from_filename(filepath)
    if fname_result:
        return fname_result, "filename"

    # ── 2-4. NER on top 10 lines ─────────────────────────────
    top = "\n".join(full_text.splitlines()[:10])

    result = get_name_bert(top, ner_pipe)
    if result:
        return result, "bert-base-top"

    result = get_name_bert(top, ner_large)
    if result:
        return result, "bert-large-top"

    result = get_name_spacy(top)
    if result:
        return result, "spacy-top"

    # ── 5. NER on full text ───────────────────────────────────
    result = get_name_bert(full_text, ner_pipe)
    if result:
        return result, "bert-base-full"

    result = get_name_bert(full_text, ner_large)
    if result:
        return result, "bert-large-full"

    result = get_name_spacy(full_text)
    if result:
        return result, "spacy-full"

    # ── 6. Regex ──────────────────────────────────────────────
    result = get_name_regex(full_text)
    if result:
        return result, "regex"

    # ── 7. Give up ────────────────────────────────────────────
    return "Name Not Found", "none"
```

```python
#  EDUCATION SECTION SPLITTER

EDUCATION_KEYWORDS = [
    "university","college","institute","school","b.tech","m.tech",
    "b.e","m.e","bsc","msc","mba","bca","mca","bachelor","master",
    "degree","diploma","10th","12th","sslc","hsc","cbse","icse",
    "engineering college","polytechnic","iit","nit","bits","vit",
    "anna university","pune university","mumbai university",
    "graduation","post graduation","pg","ug","phd","doctorate",
]

def split_experience_education(text):
    lines = text.splitlines()
    edu_headers = re.compile(
        r'^\s*(education|academic|qualification|scholastic|'
        r'educational background|academic background)\s*$', re.IGNORECASE)
    edu_start = None
    for i, line in enumerate(lines):
        if edu_headers.match(line):
            edu_start = i; break
    if edu_start is None:
        split_point = int(len(lines) * 0.70)
        return "\n".join(lines[:split_point]), "\n".join(lines[split_point:])
    return "\n".join(lines[:edu_start]), "\n".join(lines[edu_start:])

def is_education_line(line):
    return any(kw in line.lower() for kw in EDUCATION_KEYWORDS)
```

```python
#  DOMAIN SWITCH DETECTION  — Hybrid (Pretrained + Hardcode)

ROLE_CATEGORY_MAP = {
    "software engineer"        : "Software Development",
    "software developer"       : "Software Development",
    "web developer"            : "Software Development",
    "frontend developer"       : "Software Development",
    "front end developer"      : "Software Development",
    "backend developer"        : "Software Development",
    "back end developer"       : "Software Development",
    "full stack developer"     : "Software Development",
    "fullstack developer"      : "Software Development",
    "react developer"          : "Software Development",
    "java developer"           : "Software Development",
    "python developer"         : "Software Development",
    ".net developer"           : "Software Development",
    "dot net developer"        : "Software Development",
    "node developer"           : "Software Development",
    "application developer"    : "Software Development",
    "sde"                      : "Software Development",
    "swe"                      : "Software Development",
    "qa engineer"              : "QA / Testing",
    "quality analyst"          : "QA / Testing",
    "quality assurance"        : "QA / Testing",
    "test engineer"            : "QA / Testing",
    "software tester"          : "QA / Testing",
    "automation tester"        : "QA / Testing",
    "manual tester"            : "QA / Testing",
    "testing engineer"         : "QA / Testing",
    "sdet"                     : "QA / Testing",
    "qa analyst"               : "QA / Testing",
    "qa lead"                  : "QA / Testing",
    "mobile developer"         : "Mobile Development",
    "android developer"        : "Mobile Development",
    "ios developer"            : "Mobile Development",
    "flutter developer"        : "Mobile Development",
    "react native developer"   : "Mobile Development",
    "data scientist"           : "Data Science / ML",
    "data analyst"             : "Data Science / ML",
    "data engineer"            : "Data Science / ML",
    "ml engineer"              : "Data Science / ML",
    "machine learning engineer": "Data Science / ML",
    "ai engineer"              : "Data Science / ML",
    "bi developer"             : "Data Science / ML",
    "business intelligence"    : "Data Science / ML",
    "devops engineer"          : "DevOps / Cloud",
    "cloud engineer"           : "DevOps / Cloud",
    "site reliability engineer": "DevOps / Cloud",
    "sre"                      : "DevOps / Cloud",
    "infrastructure engineer"  : "DevOps / Cloud",
    "platform engineer"        : "DevOps / Cloud",
    "systems engineer"         : "DevOps / Cloud",
    "ui designer"              : "UI/UX Design",
    "ux designer"              : "UI/UX Design",
    "ui/ux designer"           : "UI/UX Design",
    "ux researcher"            : "UI/UX Design",
    "product designer"         : "UI/UX Design",
    "graphic designer"         : "UI/UX Design",
    "visual designer"          : "UI/UX Design",
    "interaction designer"     : "UI/UX Design",
    "product manager"          : "Product Management",
    "product owner"            : "Product Management",
    "business analyst"         : "Business Analysis",
    "functional analyst"       : "Business Analysis",
    "system analyst"           : "Business Analysis",
    "project manager"          : "Project Management",
    "program manager"          : "Project Management",
    "scrum master"             : "Project Management",
    "delivery manager"         : "Project Management",
    "hr executive"             : "Human Resources",
    "hr manager"               : "Human Resources",
    "recruiter"                : "Human Resources",
    "talent acquisition"       : "Human Resources",
    "sales executive"          : "Sales / Marketing",
    "sales manager"            : "Sales / Marketing",
    "marketing executive"      : "Sales / Marketing",
    "digital marketing"        : "Sales / Marketing",
    "seo analyst"              : "Sales / Marketing",
    "content writer"           : "Technical Writing",
    "technical writer"         : "Technical Writing",
    "accountant"               : "Finance / Accounting",
    "financial analyst"        : "Finance / Accounting",
    "finance manager"          : "Finance / Accounting",
    "security engineer"        : "Cybersecurity",
    "cybersecurity engineer"   : "Cybersecurity",
    "penetration tester"       : "Cybersecurity",
    "information security"     : "Cybersecurity",
    "network engineer"         : "Network Engineering",
    "network administrator"    : "Network Engineering",
    "system administrator"     : "Network Engineering",
}

DOMAIN_LABELS = [
    "Software Development", "QA / Testing", "Data Science / ML",
    "DevOps / Cloud", "UI/UX Design", "Mobile Development",
    "Product Management", "Business Analysis", "Project Management",
    "Human Resources", "Sales / Marketing", "Finance / Accounting",
    "Cybersecurity", "Network Engineering", "Technical Writing",
]

MODEL_CONFIDENCE_THRESHOLD = 0.45

DATE_RANGE_RE = re.compile(
    r'(?:jan(?:uary)?|feb(?:ruary)?|mar(?:ch)?|apr(?:il)?|may|jun(?:e)?|'
    r'jul(?:y)?|aug(?:ust)?|sep(?:tember)?|oct(?:ober)?|nov(?:ember)?|dec(?:ember)?)'
    r'\s+\d{4}'
    r'|\d{4}\s*[-–]\s*(?:\d{4}|present|current|till\s+date)',
    re.IGNORECASE
)

JOB_TITLE_LINE_RE = re.compile(
    r'(?:^|\n)\s*'
    r'((?:senior|junior|lead|principal|associate|sr\.?|jr\.?|staff|assistant)?\s*'
    r'(?:'
    r'software\s+(?:engineer|developer)|'
    r'(?:frontend|front[\s\-]end|backend|back[\s\-]end|full[\s\-]?stack)\s+developer|'
    r'(?:mobile|android|ios|flutter|react\s+native)\s+developer|'
    r'(?:qa|quality\s+assurance|quality\s+analyst|test(?:ing)?)\s+(?:engineer|analyst|lead)?|'
    r'software\s+tester|automation\s+tester|manual\s+tester|sdet|'
    r'data\s+(?:scientist|analyst|engineer)|'
    r'(?:machine\s+learning|ml|ai)\s+engineer|'
    r'(?:devops|cloud|platform|infrastructure|systems?)\s+engineer|'
    r'site\s+reliability\s+engineer|sre|'
    r'(?:ui[\s/]*ux|product|graphic|visual|interaction)\s+designer|'
    r'product\s+(?:manager|owner)|'
    r'business\s+analyst|project\s+manager|scrum\s+master|'
    r'(?:hr|human\s+resource)\s+(?:manager|executive)|recruiter|'
    r'(?:sales|marketing)\s+(?:manager|executive)|'
    r'(?:network|security|cybersecurity)\s+engineer|'
    r'technical\s+writer|financial\s+analyst|accountant|'
    r'sde|swe'
    r'))',
    re.IGNORECASE | re.MULTILINE
)

def hardcode_category_from_title(title):
    title_lower = title.lower().strip()
    for keyword in sorted(ROLE_CATEGORY_MAP.keys(), key=len, reverse=True):
        if keyword in title_lower:
            return ROLE_CATEGORY_MAP[keyword]
    return None

def model_category_from_block(block_text):
    try:
        result = classifier(block_text[:600], DOMAIN_LABELS, multi_label=False)
        return result["labels"][0], result["scores"][0]
    except:
        return None, 0.0

def extract_title_from_block(block_text):
    match = JOB_TITLE_LINE_RE.search(block_text)
    if match: return match.group(1).strip()
    return None

def resolve_category_hybrid(block_text):
    """
    1. Run pretrained model → model_cat, confidence
    2. Extract job title → hardcode_cat
    3. Conflict resolution:
       - Both agree              → trust both
       - Conflict                → hardcode wins (title is explicit)
       - Only hardcode           → use hardcode
       - Model confident only    → use model
       - Model low-conf only     → use model (best available)
    """
    model_cat, confidence = model_category_from_block(block_text)
    title                 = extract_title_from_block(block_text)
    hardcode_cat          = hardcode_category_from_title(title) if title else None

    if model_cat and hardcode_cat and model_cat == hardcode_cat:
        return model_cat, f"model+hardcode agree ({confidence:.0%})"
    if model_cat and hardcode_cat and model_cat != hardcode_cat:
        return hardcode_cat, f"hardcode override (model={model_cat} {confidence:.0%})"
    if hardcode_cat and not model_cat:
        return hardcode_cat, "hardcode only"
    if model_cat and confidence >= MODEL_CONFIDENCE_THRESHOLD and not hardcode_cat:
        return model_cat, f"model only ({confidence:.0%})"
    if model_cat and not hardcode_cat:
        return model_cat, f"model low-conf ({confidence:.0%})"
    return "Unknown", "no match"

def split_into_job_blocks(experience_text):
    lines, blocks, current = experience_text.splitlines(), [], []
    for line in lines:
        if DATE_RANGE_RE.search(line) and current:
            blocks.append("\n".join(current))
            current = [line]
        else:
            current.append(line)
    if current: blocks.append("\n".join(current))
    return [b for b in blocks if len(b.strip()) > 40]

def detect_domains(full_text):
    """
    Returns: (current_domain, previous_domain_or_None, switched: bool)
    switched=True  only when job ROLE category genuinely changed
    e.g. Software Developer → QA Tester = switched
         Developer at Company A → Developer at Company B = NOT switched
    """
    experience_text, _ = split_experience_education(full_text)
    blocks = split_into_job_blocks(experience_text)

    if not blocks:
        cat, conf = model_category_from_block(experience_text[:1200] or full_text[:1200])
        return cat or "Unknown", None, False

    block_categories = []
    for block in blocks:
        cat, _ = resolve_category_hybrid(block)
        if cat and cat != "Unknown":
            block_categories.append(cat)

    if not block_categories:
        return "Unknown", None, False

    # Remove consecutive duplicates
    deduped = []
    for cat in block_categories:
        if not deduped or deduped[-1] != cat:
            deduped.append(cat)

    current_domain = deduped[-1]

    if len(deduped) >= 2:
        previous_domain = None
        for cat in deduped[:-1]:
            if cat != current_domain:
                previous_domain = cat
                break
        if previous_domain:
            return current_domain, previous_domain, True

    return current_domain, None, False
```

```python
#  CAREER BREAK DETECTION  (with exact date ranges)


MONTH_MAP = {
    "jan":1,"feb":2,"mar":3,"apr":4,"may":5,"jun":6,
    "jul":7,"aug":8,"sep":9,"oct":10,"nov":11,"dec":12,
    "january":1,"february":2,"march":3,"april":4,"june":6,
    "july":7,"august":8,"september":9,"october":10,
    "november":11,"december":12
}
MONTH_NAMES = {
    1:"Jan",2:"Feb",3:"Mar",4:"Apr",5:"May",6:"Jun",
    7:"Jul",8:"Aug",9:"Sep",10:"Oct",11:"Nov",12:"Dec"
}

def parse_date(token):
    token = token.strip().lower()
    if token in ("present","current","till date","now"):
        now = datetime.now(); return now.month, now.year
    m = re.match(r'([a-z]+)\s+(\d{4})', token)
    if m:
        mon = MONTH_MAP.get(m.group(1)); yr = int(m.group(2))
        if mon: return mon, yr
    m = re.match(r'(\d{4})$', token)
    if m: return 6, int(m.group(1))
    return None

def extract_date_ranges(text):
    pattern = re.compile(
        r'((?:jan(?:uary)?|feb(?:ruary)?|mar(?:ch)?|apr(?:il)?|may|jun(?:e)?|'
        r'jul(?:y)?|aug(?:ust)?|sep(?:tember)?|oct(?:ober)?|nov(?:ember)?|dec(?:ember)?)'
        r'\s+\d{4}|\d{4})'
        r'\s*(?:–|-|to)\s*'
        r'((?:jan(?:uary)?|feb(?:ruary)?|mar(?:ch)?|apr(?:il)?|may|jun(?:e)?|'
        r'jul(?:y)?|aug(?:ust)?|sep(?:tember)?|oct(?:ober)?|nov(?:ember)?|dec(?:ember)?)'
        r'\s+\d{4}|\d{4}|present|current|till date)',
        re.IGNORECASE
    )
    ranges = []
    for match in pattern.finditer(text):
        s = parse_date(match.group(1)); e = parse_date(match.group(2))
        if s and e:
            start_m = s[1]*12 + s[0]; end_m = e[1]*12 + e[0]
            if end_m >= start_m:
                ranges.append((start_m, end_m, s, e))
    return sorted(ranges, key=lambda x: x[0])

def abs_to_monthyear(abs_m):
    yr = (abs_m-1)//12; mon = abs_m - yr*12; return mon, yr

def detect_career_break(text):
    experience_text, _ = split_experience_education(text)
    ranges = extract_date_ranges(experience_text)
    if len(ranges) < 2: return False, "No career break detected"
    breaks = []
    for i in range(1, len(ranges)):
        prev_end_abs   = ranges[i-1][1]
        curr_start_abs = ranges[i][0]
        gap            = curr_start_abs - prev_end_abs
        if gap >= 6:
            bsm, bsy = abs_to_monthyear(prev_end_abs)
            bem, bey = abs_to_monthyear(curr_start_abs)
            start_label = f"{MONTH_NAMES.get(bsm,'?')} {bsy}"
            end_label   = f"{MONTH_NAMES.get(bem,'?')} {bey}"
            yrs = gap//12; mons = gap%12
            if yrs > 0 and mons > 0: dur = f"{yrs}yr {mons}mo"
            elif yrs > 0:             dur = f"{yrs}yr{'s' if yrs>1 else ''}"
            else:                     dur = f"{mons}mo{'s' if mons>1 else ''}"
            breaks.append(f"{start_label} → {end_label} ({dur})")
    if breaks:
        return True, "Career break: " + "  |  ".join(breaks)
    return False, "No career break detected"
```

```python
#  MATCH SCORE

def compute_match_score(jd_text, resume_text):
    jd_emb  = embedder.encode(jd_text,            convert_to_tensor=True)
    res_emb = embedder.encode(resume_text[:3000], convert_to_tensor=True)
    return round(util.cos_sim(jd_emb, res_emb).item() * 100, 2)
```

```python
#  BUILD SUMMARY


def build_summary(name, method, current_d, prev_d, switched,
                  has_break, break_desc, location_str):
    lines = []
    lines.append(f"Candidate     : {name}  (via {method})")
    lines.append(f"Current Domain: {current_d}")
    if switched:
        lines.append(f"Domain Switch : Yes  ({prev_d}  →  {current_d})  ⚠️")
    else:
        lines.append(f"Domain Switch : No")
    if has_break:
        lines.append(f"Career Break  : ⚠️ {break_desc}")
    else:
        lines.append(f"Career Break  : No career break detected")
    lines.append(f"Locations     : {location_str}")
    return "\n".join(lines)
```

```python
#  ANALYSE A SINGLE RESUME

def analyze_resume(jd_text, path, threshold=70):
    fname                          = os.path.basename(path)
    full_text, left_col, right_col = extract_text(path)

    if not full_text.strip():
        return {
            "file": fname, "score": 0, "status": "ERROR",
            "name": "N/A", "domain": "—",
            "domain_switch": False,
            "domain_switch_label": "No",
            "domain_switch_from": "",
            "career_break": False,
            "career_break_detail": "—",
            "locations": "Location not given",
            "summary": "Could not extract text from file."
        }

    name, method                    = extract_name(full_text, left_col, right_col, path)
    score                           = compute_match_score(jd_text, full_text)
    status                          = "MATCHED" if score >= threshold else "NOT MATCHED"
    current_d, prev_d, switched     = detect_domains(full_text)
    has_break, break_desc           = detect_career_break(full_text)
    location_str                    = extract_company_locations(full_text, left_col, right_col)

    switch_label = f"Yes  ({prev_d}  →  {current_d})" if switched else "No"

    summary = build_summary(
        name, method, current_d, prev_d, switched,
        has_break, break_desc, location_str
    )

    return {
        "file"               : fname,
        "score"              : score,
        "status"             : status,
        "name"               : name,
        "domain"             : current_d,
        "domain_switch"      : switched,
        "domain_switch_label": switch_label,
        "domain_switch_from" : prev_d if switched else "",
        "career_break"       : has_break,
        "career_break_detail": break_desc,
        "locations"          : location_str,
        "summary"            : summary
    }
```

```python
#  DISPLAY RESULTS

def display_results(results, threshold):
    clear_output(wait=True)
    print(f"\n{'='*75}")
    print(f"   RESUME MATCHING REPORT  |  Threshold: {threshold}%")
    print(f"{'='*75}\n")

    matched = sum(1 for r in results if "NOT" not in r["status"] and "MATCHED" in r["status"])
    print(f"Total: {len(results)}  |   Matched: {matched}  |   Not Matched: {len(results)-matched}\n")
    print("─"*75)

    for r in sorted(results, key=lambda x: x.get("score", 0), reverse=True):
        print(f"\n  File        : {r['file']}")
        print(f"  Match Score : {r['score']}%  →  {r['status']}")
        print(f"\n  ── Summary ──────────────────────────────────────────────────")
        for line in r["summary"].splitlines():
            print(f"  {line}")
        print("─"*75)

    # ── DataFrame summary table ──
    rows = []
    for r in results:
        # Parse name from summary for table
        name_line = next((l for l in r["summary"].splitlines() if "Candidate" in l), "")
        name_val  = name_line.split(":", 1)[-1].split("(via")[0].strip() if name_line else "—"

        rows.append({
            "File"       : r["file"],
            "Name"       : name_val,
            "Score (%)"  : r["score"],
            "Status"     : r["status"],
            "Summary"    : r["summary"].replace("\n", " | ")
        })

    df = pd.DataFrame(rows)
    print("\n SUMMARY TABLE:\n")
    display(df.sort_values("Score (%)", ascending=False).reset_index(drop=True))
```

```python
app = Flask(__name__)

@app.route("/api/health", methods=["GET"])
def health():
    return jsonify({"status": "ok", "message": "Resume Matcher API is running"})


@app.route("/api/analyze", methods=["POST"])
def analyze_single():
    if "file" not in request.files:
        return jsonify({"error": "No file uploaded"}), 400
    if not request.form.get("jd_text", "").strip():
        return jsonify({"error": "jd_text is required"}), 400

    resume_file = request.files["file"]
    jd_text     = request.form["jd_text"]
    threshold   = int(request.form.get("threshold", 70))

    suffix = os.path.splitext(resume_file.filename)[-1]
    with tempfile.NamedTemporaryFile(delete=False, suffix=suffix) as tmp:
        resume_file.save(tmp.name)
        tmp_path = tmp.name

    try:
        result = analyze_resume(jd_text, tmp_path, threshold)
        result["file"] = resume_file.filename
        return jsonify(result)
    except Exception as e:
        return jsonify({"error": str(e)}), 500
    finally:
        os.unlink(tmp_path)


@app.route("/api/analyze-batch", methods=["POST"])
def analyze_batch():
    files     = request.files.getlist("files")
    jd_text   = request.form.get("jd_text", "")
    threshold = int(request.form.get("threshold", 70))

    if not files:
        return jsonify({"error": "No files uploaded"}), 400
    if not jd_text.strip():
        return jsonify({"error": "jd_text is required"}), 400

    results = []
    for resume_file in files:
        suffix = os.path.splitext(resume_file.filename)[-1]
        with tempfile.NamedTemporaryFile(delete=False, suffix=suffix) as tmp:
            resume_file.save(tmp.name)
            tmp_path = tmp.name
        try:
            result = analyze_resume(jd_text, tmp_path, threshold)
            result["file"] = resume_file.filename
            results.append(result)
        except Exception as e:
            results.append({"file": resume_file.filename, "error": str(e)})
        finally:
            os.unlink(tmp_path)

    matched     = sum(1 for r in results if r.get("status") == "MATCHED")
    not_matched = len(results) - matched

    return jsonify({
        "total"      : len(results),
        "matched"    : matched,
        "not_matched": not_matched,
        "threshold"  : threshold,
        "results"    : sorted(results, key=lambda x: x.get("score", 0), reverse=True)
    })
```
